# Pyber Power of Plots

* Examining the scatter plot, more densely populated urban areas have both a higher driver count and a higher total number of rides, which is not surprising. The data also indicates that this situation may foster competition, which drives down prices, since the average fare is lower in these areas. 


* The pie charts show a similar portion share of rides and fares between urban, suburban, and rural cities. However, the share for total drivers in suburban cities is a bit smaller, and in rural cities significantly smaller. This indicates that there may be opportunity for additional drivers in those areas.


* There is more variation, or range, in the average fare among rural cities than the other two groups. This is likely a result of a wider range of distance covered by rides. Even in very rural areas, customers may only need to go a short distance, but they are more likely to need to cover a longer distance.


```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# Read the Mouse and Drug Data and the Clinical Trial Data
city_df = pd.read_csv("data/city_data.csv")
ride_df = pd.read_csv("data/ride_data.csv")

# Combine the data into a single dataset
city_ride_df= pd.merge(city_df, ride_df, how="left", on=["city"])

#Reorder and sort columns
city_ride_df = city_ride_df[["city","date","fare","ride_id","driver_count","type"]]
city_ride_df = city_ride_df.sort_values("ride_id")

# Display the data table for preview
city_ride_df.head()
```

## Bubble Plot of Ride Sharing Data

```python
urban_city = city_ride_df.loc[(city_ride_df["type"] == "Urban")]
suburban_city = city_ride_df.loc[(city_ride_df["type"] == "Suburban")]
rural_city = city_ride_df.loc[(city_ride_df["type"] == "Rural")]
```

```python
# Per city calculations of Average Fare, Total Rides and Total Drivers for Urban cities
avg_fare_urban_city = urban_city.groupby(['city'])['fare'].mean()
total_rides_urban_city = urban_city.groupby(['city']).count()['ride_id']
total_drivers_urban_city = urban_city.groupby(['city'])['driver_count'].value_counts()

# Per city calculations of Average Fare, Total Rides and Total Drivers for Suburban cities
avg_fare_suburban_city = suburban_city.groupby(['city'])['fare'].mean()
total_rides_suburban_city = suburban_city.groupby(['city']).count()['ride_id']
total_drivers_suburban_city = suburban_city.groupby(['city'])['driver_count'].value_counts()

# Per city calculations of Average Fare, Total Rides and Total Drivers for Rural cities
avg_fare_rural_city = rural_city.groupby(['city'])['fare'].mean()
total_rides_rural_city = rural_city.groupby(['city']).count()['ride_id']
total_drivers_rural_city = rural_city.groupby(['city'])['driver_count'].value_counts()
```

```python
# Create scatter plots for Urban, Suburban and Rural cities
plt.scatter(total_rides_urban_city, avg_fare_urban_city,s=total_drivers_urban_city*20,
            marker ='o', facecolors ="lightcoral", edgecolors='black',alpha = 0.75, label="Urban")

plt.scatter(total_rides_suburban_city, avg_fare_suburban_city,s=total_drivers_suburban_city*20,
            marker ='o', facecolors ="lightskyblue", edgecolors='black',alpha = 0.75, label="Suburban")

plt.scatter(total_rides_rural_city, avg_fare_rural_city,s=total_drivers_rural_city*20,
            marker ='o', facecolors ="gold", edgecolors='black',alpha = 0.75, label="Rural")

plt.legend(loc="upper right")

plt.title("Pyber Ride Sharing Data (2018)")
plt.xlabel("Total Number of Rides (Per City)")
plt.ylabel("Average Fare ($)")
plt.grid()
plt.tight_layout()
plt.savefig("./pyber_bubble_plot.png")
plt.figtext(1, .2, "Size indicates driver count")
plt.show()
```

## Total Fares by City Type

```python
# Group fare by type
city_fare_pie = city_ride_df.groupby(["type"])["fare"].sum()
```

```python
colors = ["gold", "lightskyblue", "lightcoral"]
explode = (0,0.1, 0.1)
plt.pie(city_fare_pie, labels = city_fare_pie.index, autopct = "%1.1f%%", 
        colors = colors, explode = explode, shadow = True, startangle = 140)
plt.title("% of Total Fares by City Type")
plt.tight_layout()
plt.savefig("./pyber_city_fare_pie.png")
plt.show()
```

## Total Rides by City Type

```python
city_rides_pie = city_ride_df.groupby(["type"])["ride_id"].count()
```

```python
colors = ["gold", "lightskyblue", "lightcoral"]
explode = (0,0.1, 0.1)
plt.pie(city_rides_pie, labels = city_rides_pie.index, autopct = "%1.1f%%", 
        colors = colors, explode = explode, shadow = True, startangle = 140)
plt.title("% of Total Rides by City Type")
plt.tight_layout()
plt.savefig("./pyber_city_rides_pie.png")
plt.show()
```

## Total Drivers by City Type

```python
city_drivers_pie = city_df.groupby(["type"])["driver_count"].sum()
```

```python
colors = ["gold", "lightskyblue", "lightcoral"]
explode = (0,0.1, 0.1)
plt.pie(city_drivers_pie, labels = city_drivers_pie.index, autopct = "%1.1f%%", 
        colors = colors, explode = explode, shadow = True, startangle = 140)
plt.title("% of Total Drivers by City Type")
plt.tight_layout()
plt.savefig("./pyber_city_fare_pie.png")
plt.show()
```
